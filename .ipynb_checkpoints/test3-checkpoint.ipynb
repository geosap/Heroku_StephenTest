```python
graph = load_graphml('Resources/Sprint.graphml')

adjacency = graph.adjacency
names = graph.names

lon = graph.node_attribute.Longitude
lat = graph.node_attribute.Latitude
position = np.stack((lon, lat), axis=1)
```

```python
graph_shortest_path(dist_matrix = position)
```

```python
fig, ax = plt.subplots()

nx.draw_networkx_nodes(G, pos = position, node_size = 10, node_color = 'red', edge_color = 'k', alpha = .5, with_label = True)
nx.draw_networkx_edges(G, pos = position, edge_color = 'gray', alpha = .1)
nx.draw_networkx_labels(G, position , label_pos = 10.3)

mplleaflet.display(fig=ax.figure)
```
